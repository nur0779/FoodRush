## Ilmu data
</div>

Dengan Colab, Anda dapat memanfaatkan kecanggihan library Python yang populer untuk menganalisis dan memvisualisasikan data. Sel kode di bawah menggunakan <strong>numpy</strong> untuk menghasilkan beberapa data acak, dan menggunakan <strong>matplotlib</strong> untuk memvisualisasikannya. Untuk mengedit kode, cukup klik sel dan mulai pengeditan.

Anda dapat mengimpor data sendiri ke notebook Colab dari akun Google Drive Anda, termasuk dari spreadsheet, GitHub, dan berbagai sumber lainnya. Untuk mempelajari lebih lanjut pengimporan data dan bagaimana Colab dapat digunakan untuk ilmu data, lihat link di bawah pada bagian <a href="#working-with-data">Bekerja dengan Data</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# --- Install Graphviz di Colab ---
!apt-get install graphviz -y
!pip install graphviz

from graphviz import Digraph
from IPython.display import Image

# ==== 1. Use Case Diagram ====
usecase = Digraph("UseCase", format="png")
usecase.attr(label="Use Case Diagram - FoodRush", labelloc="t")

usecase.node("User", shape="box")
cases = [
    "Login / Sign Up", "Melihat Menu", "Add to Cart",
    "Lihat Cart", "Edit Quantity / Hapus Item",
    "Checkout", "Kirim Pesan (About)"
]
for c in cases:
    usecase.node(c, shape="ellipse")
    usecase.edge("User", c)

usecase.render("uml_usecase_foodrush", cleanup=True)
display(Image("uml_usecase_foodrush.png"))

# ==== 2. Activity Diagram (Add to Cart) ====
activity = Digraph("Activity", format="png")
activity.attr(label="Activity Diagram - Add to Cart", labelloc="t")

activity.node("Start", shape="circle")
activity.node("Pilih", "User pilih produk", shape="box")
activity.node("Klik", "Klik tombol Add to Cart", shape="box")
activity.node("Ambil", "Ambil data cart dari localStorage", shape="box")
activity.node("Check", "Produk sudah ada di cart?", shape="diamond")
activity.node("TambahQty", "Tambah quantity +1", shape="box")
activity.node("TambahBaru", "Tambah produk baru", shape="box")
activity.node("Simpan", "Simpan cart ke localStorage", shape="box")
activity.node("Notif", "Tampilkan notifikasi", shape="box")
activity.node("End", shape="doublecircle")

activity.edges([("Start","Pilih"),("Pilih","Klik"),("Klik","Ambil"),("Ambil","Check")])
activity.edge("Check","TambahQty",label="Ya")
activity.edge("Check","TambahBaru",label="Tidak")
activity.edge("TambahQty","Simpan")
activity.edge("TambahBaru","Simpan")
activity.edge("Simpan","Notif")
activity.edge("Notif","End")

activity.render("uml_activity_foodrush", cleanup=True)
display(Image("uml_activity_foodrush.png"))

# ==== 3. Class Diagram ====
classd = Digraph("Class", format="png")
classd.attr(label="Class Diagram - FoodRush", labelloc="t")

classd.node("UserClass", """{
User|
- nama
- email
- password|
+ login()
+ logout()
+ signUp()
}""", shape="record")

classd.node("ItemClass", """{
Item|
- name
- price
- quantity|
+ getSubtotal()
}""", shape="record")

classd.node("CartClass", """{
Cart|
- items[]|
+ addItem()
+ removeItem()
+ updateQty()
+ getTotal()
}""", shape="record")

classd.node("TransactionClass", """{
Transaction|
- user
- cart
- alamat
- metodeBayar|
+ checkout()
}""", shape="record")

classd.edge("UserClass","CartClass",label="has")
classd.edge("CartClass","ItemClass",label="contains")
classd.edge("UserClass","TransactionClass",label="creates")
classd.edge("TransactionClass","CartClass",label="uses")

classd.render("uml_class_foodrush", cleanup=True)
display(Image("uml_class_foodrush.png"))

# ==== 4. Object Diagram ====
objectd = Digraph("Object", format="png")
objectd.attr(label="Object Diagram - FoodRush", labelloc="t")

objectd.node("UserObj", "{User: Moch Nur | email= nur@gmail.com}", shape="record")
objectd.node("Item1", "{Item: Donat | price=10000 | qty=2}", shape="record")
objectd.node("Item2", "{Item: Pizza | price=20000 | qty=1}", shape="record")
objectd.node("CartObj", "{Cart | items=[Donat, Pizza]}", shape="record")
objectd.node("TransObj", "{Transaction | alamat=Jl. Raya Surabaya | metode=COD}", shape="record")

objectd.edge("UserObj","CartObj")
objectd.edge("CartObj","Item1")
objectd.edge("CartObj","Item2")
objectd.edge("UserObj","TransObj")
objectd.edge("TransObj","CartObj")

objectd.render("uml_object_foodrush", cleanup=True)
display(Image("uml_object_foodrush.png"))

# ==== 5. Sequence Diagram (Add to Cart) ====
seq = Digraph("Sequence", format="png")
seq.attr(rankdir="LR", label="Sequence Diagram - Add to Cart", labelloc="t")

actors = ["User","UI","Cart","LocalStorage"]
for a in actors:
    seq.node(a, a, shape="box")

seq.edge("User","UI","Pilih produk & klik Add to Cart")
seq.edge("UI","Cart","addItem()")
seq.edge("Cart","LocalStorage","Ambil data cart")
seq.edge("LocalStorage","Cart","Return cart data")
seq.edge("Cart","LocalStorage","Update cart & Simpan")
seq.edge("UI","User","Tampilkan notifikasi")

seq.render("uml_sequence_foodrush", cleanup=True)
display(Image("uml_sequence_foodrush.png"))

# ==== 6. Communication Diagram ====
comm = Digraph("Comm", format="png")
comm.attr(label="Communication Diagram - Checkout", labelloc="t")

comm.node("User","User",shape="ellipse")
comm.node("Form","Checkout Form",shape="box")
comm.node("Transaction","Transaction",shape="box")
comm.node("Cart","Cart",shape="box")
comm.node("System","System Notif",shape="box")

comm.edge("User","Form","Isi data")
comm.edge("Form","Transaction","Submit")
comm.edge("Transaction","Cart","Ambil data cart")
comm.edge("Transaction","System","Simpan transaksi")
comm.edge("System","User","Tampilkan notifikasi")

comm.render("uml_comm_foodrush", cleanup=True)
display(Image("uml_comm_foodrush.png"))
```
